df.duplicated() returns a boolean Series where True indicates that the row is a duplicate.
df[df.duplicated()] filters the DataFrame to show only the duplicate rows.
The print() statement displays these duplicate rows.

```python
# Check for duplicate rows in the entire DataFrame
data.duplicated()
# if true then it is duplicated row
duplicates = data[data.duplicated()]
duplicates
# Print the duplicate rows
```

If you want to check for duplicates based on specific columns (e.g., only by Name and Age), 
you can pass the column names to the duplicated() method like this:

# Check for duplicates based on specific columns
duplicates = df[df.duplicated(subset=['Name', 'Age'])]

# Print the duplicate rows based on specific columns
print("Duplicate Rows based on Name and Age:")
print(duplicates)
